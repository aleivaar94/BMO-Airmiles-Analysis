## Loading and prepping data with all personal expenses

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

expenses = pd.read_csv('expenses.csv')
expenses.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11'], axis=1, inplace=True)
expenses.rename(columns={'Category (Select)':'Category', 'Date (mm/dd/yyyy)':'Date',
                        'Subcategory (Select)':'Subcategory', 'credit card':'credit_card'}, inplace=True)
expenses.columns = expenses.columns.str.strip().str.lower()

# Replacing NaN values
expenses.loc[609,'year'] = 2021
expenses.loc[609,'month'] = 'September'
expenses.loc[319,'credit card'] = 9880

# Replacing values to change to data type 'float'
expenses['amount'] = expenses.amount.str.replace('$','', regex=True).str.replace(',','', regex=True).str.replace(' -   ','', regex=True)
expenses['credit card'] = expenses['credit card'].str.replace('Cash','9880', regex=True)

# Dropping empty ('') values to allow data type change to 'float'. Only lost 2 rows of data
expenses.drop(expenses[expenses['amount'] == ''].index, inplace=True)

# Changing data types for easier analysis
expenses['year'] = expenses.year.astype('int')
expenses['amount'] = expenses.amount.astype('float')
expenses['date'] = pd.to_datetime(expenses.date, format='%m/%d/%y')

# expenses['month'] = pd.to_datetime(expenses.month, format='%B').dt.month
# expenses.head()
```

## Calculating Air Miles Earned from 2021 Credit Card Expenses
##### Excluding Safeway (affiliated air miles grocery store)

Air Miles are awarded depending on the amount spent. The normal rate is 1 mile for every \$20.0 dollars. However, if you use a credit card from a bank affiliated with Air Miles you can earn more Air Miles with less money spent.

- Standard Air Miles: 1 miles for every \$20.0 spent.
- BMO Air Miles Mastercard: 3 miles for every \$25.0 spent.
- BMO Air Miles World Elite: 3 miles for every \$12.0 spent.

95 miles = \$10.0

or

9.5 miles = \$1.0

```python
# pd.set_option('display.max_rows', None)
expenses_2021 = expenses[(expenses['year'] == 2021) & (expenses['store'] != 'Safeway') 
                         & (expenses['subcategory'] != 'Rent') & (expenses['credit card'] != '9880')]
expenses_2021['amount'] = expenses_2021['amount'].apply(np.floor)
expenses_2021.reset_index(drop=True, inplace=True)
expenses_2021['miles'] = (expenses_2021.amount//20)
expenses_2021['bmo_miles'] = 3*(expenses_2021.amount//25)
expenses_2021['bmo_elite_miles'] = 3*(expenses_2021.amount//12)
expenses_2021.head()
```

##### Estimating Air Miles and Cash that could have been earned in 2021

```python
total_miles_2021 = expenses_2021.miles.sum()
bmo_miles_2021 = expenses_2021.bmo_miles.sum()
bmo_elite_miles_2021 = expenses_2021.bmo_elite_miles.sum()
print('Total miles: ' + str(round(total_miles_2021,2)) + '\n' + 'Cash: ' + str(round(total_miles_2021/9.5,2)))
print('BMO miles: ' + str(round(bmo_miles_2021,2)) + '\n' + 'Cash: ' + str(round(bmo_miles_2021/9.5, 2)))
print('BMO Elite miles: ' + str(round(bmo_elite_miles_2021,2))  + '\n' + 'Cash: ' + str(round(bmo_elite_miles_2021/9.5, 2)))
```

## Calculating Air Miles earned exclusively in Safeway

Because Safeway is an affiliated grocery store, one can earn 3x the miles if one has a BMO or BMO Elite miles credit card. Since I know that I do most of my groceries at Safeway and that Safeway is the main store from which I earn my Air Miles, I calculate how many air miles I earn exclusively from this store.

```python
safeway = expenses[(expenses['store'] == 'Safeway') & (expenses['credit card'] != '9880')]
safeway.reset_index(drop=True, inplace=True)

# Round down to the nearest whole number
safeway['amount'] = safeway['amount'].apply(np.floor)
safeway['miles'] = (safeway.amount//20)
safeway['bmo_miles'] = 3*(safeway.amount//25)
safeway['bmo_elite_miles'] = 3*(safeway.amount//12)
# Alternative to using floor division
# safeway['miles_earned'] = pd.cut(safeway.amount, range(20,121,20), labels = range(1,6), right = False)
# safeway['miles_earned'] = safeway['miles_earned'].astype(np.float)
safeway.head()
```

##### Estimating Air Miles and Cash *earned* on Safeway from groceries made from 2020 to 2022

```python
safeway_miles = safeway.miles.sum()
safeway_bmo_miles = safeway.bmo_miles.sum()
safeway_bmo_elite_miles = safeway.bmo_elite_miles.sum()

print('Total miles: ' + str(round(safeway_miles,2)) + '\n' + 'Cash: ' + str(round(safeway_miles/9.5,2)))
print('BMO miles: ' + str(round(safeway_bmo_miles,2)) + '\n' + 'Cash: ' + str(round(safeway_bmo_miles/9.5, 2)))
print('BMO Elite miles: ' + str(round(safeway_bmo_miles,2))  + '\n' + 'Cash: ' + str(round(safeway_bmo_miles/9.5, 2)))
```

## Analyzing Safeway Spending
#### How much do I spend per month on groceries?

```python
safeway_grouped = safeway.groupby(['year','month']).agg(amount_sum=('amount', 'sum'), amount_std=('amount','std')).reset_index()
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
safeway_grouped['month'] = pd.Categorical(safeway_grouped['month'], categories=months, ordered=True)
safeway_grouped.sort_values(by=['year', 'month'], inplace=True)
safeway_grouped.reset_index(drop=True).head(20)
```

#### Most of my purchases at Safeway are less than \$20.0 dollars, but I make about 5 purchasees per month, on average (plus/minus 3 purchases).

```python
safeway_grouped = safeway.groupby(['year','month']).agg(amount_sum=('amount', 'sum'), amount_std=('amount','std'))
safeway_grouped['qty_purchases'] = safeway.groupby(['year','month']).size()
safeway_grouped.reset_index(inplace=True)
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
safeway_grouped['month'] = pd.Categorical(safeway_grouped['month'], categories=months, ordered=True)
safeway_grouped.sort_values(by=['year', 'month'], inplace=True)
safeway_grouped.reset_index(drop=True)
```

```python
print('Avg. ' + str(round(safeway_grouped.qty_purchases.mean(),2)))
print('Std. ' + str(round(safeway_grouped.qty_purchases.std(),2)))
```

## Summary Table of Miles Earned
I have a Scotiabank cash back credit card that has given me \$220.0 dollars in cash back for all my expenses paid with this credit card as of August 2022. Is the BMO Air Miles credit card offer better than the Scotiabank credit card I have rigth now?

```python
summary_miles = pd.DataFrame({'2021':{'miles_cash':total_miles_2021/9.5, 
                              'bmo_cash':bmo_miles_2021/9.5,
                              'bmo_elite_cash': bmo_elite_miles_2021/9.5},
                              '2022':{'miles_cash':total_miles_2022/9.5, 
                              'bmo_cash':bmo_miles_2022/9.5,
                              'bmo_elite_cash': bmo_elite_miles_2022/9.5},
                              'safeway':{'miles_cash':safeway_miles/9.5, 
                              'bmo_cash':safeway_bmo_miles/9.5,
                              'bmo_elite_cash': safeway_bmo_elite_miles/9.5},
                             })

summary_miles['total_2021'] = summary_miles['2021'] + summary_miles.safeway
summary_miles['total_2022'] = summary_miles['2022'] + summary_miles.safeway
# summary_miles['diff_2021'] = summary_miles.total_2021 - 110
# summary_miles['diff_2022'] = summary_miles.total_2022 - 110

# $39.0 is the annual fee of Scotiabank cash back credit card
summary_miles['scotia_diff_2021'] = summary_miles.total_2021 - (220-39)
summary_miles['scotia_diff_2022'] = summary_miles.total_2022 - (220-39)

# $120.0 is the annueal fee of BMO Air Miles Mastercard credit card
summary_miles.loc['bmo_elite_cash'] =  summary_miles.loc['bmo_elite_cash'] - 120

summary_miles.round(1)
```

```python
list(summary_miles.columns)
cols = ['total_2021','total_2022', 'scotia_diff_2021','scotia_diff_2022']
```

```python
summary_miles = summary_miles[cols]
```

```python
summary_miles.round(1)
```

Based on the table above, the BMO Air Miles World Elite credit card is better than the Scotiabank cash back credit card I currently have. With the BMO Elite card I could have earned almost \\$300.0 dollars worth for miles or \\$111.0 dollars more than my current card. 

In 2022 the BMO Elite card still doesn't bring me benefit. In fact, I would have earned \\$167 instead of \\$181 dollars currently earned with my Scotia card. However, the year is not over yet, and I am sure I will accumulate more purchases which will make the BMO Elite card bring me more cash back by the end of the year.
